# 欠損値・外れ値

```python
#全データから確認
category_paths = glob.glob("../data/*")
print("全データ")
print(category_paths)

not_category = ["../data/CHANGES.txt", "../data/README.txt"]
category_paths = [f for f in category_paths if f not in not_category]
print("カテゴリのみ")
print(category_paths)
```

```python
#　カテゴリ別データ
all_category = {} # カテゴリ名：　[1news, 1news]
for path in category_paths:
    *_, category = path.split("/")
    all_category[category] = []

one_category = [] # [1news, 1news]
for category in all_category.keys():
    one_category_paths = glob.glob(f"../data/{category}/{category}*.txt")
    all_category[category] = one_category_paths
    
print("カテゴリ別データ数")
for category, paths in all_category.items():
    print(len(paths), category)
```

```python
# カテゴリ別のサンプル
for category, paths in all_category.items():
    with open(paths[0]) as f:
        text = f.readlines()
    text = "".join(text)

    print("="*50)
    print(category)
    print(text)
```

**確認**
- データ数は500~900テキスト
- 多くが700ファイル以上ある
- カテゴリ名から`peachy` `smax`のイメージがつかない
- 最初の2行は`URL` `投稿日時`のルール

```python
# テキスト数
category_min_len = {}
category_max_len = {}
for category, paths in all_category.items():
    for i, path in enumerate(paths):
        with open(path) as f:
            text = f.readlines()
        text = "".join(text[2:])

        if i == 0:
            text_min_len = len(text)
            text_max_len = len(text)
        else:
            if len(text) < text_min_len:
                text_min_len = len(text)
            if len(text) > text_max_len:
                text_max_len = len(text)

    category_min_len[category] = text_min_len
    category_max_len[category] = text_max_len   
```

```python
category_min_len
```

```python
category_max_len
```

**確認**
- 最小テキスト数は、100~500字程度 →　100字の中にはURLや日時も含んでいるため、実際のデータ量はかなり少ないと判断
- 最大テキスト数は、2000~12000字程度　→　`it-life-hack` `smax`は文字数が倍以上多いのが存在

```python
# 最小テキストの確認
for category, paths in all_category.items():
    min_len = category_min_len[category]
    for i, path in enumerate(paths):
        with open(path) as f:
            text = f.readlines()
        # URLと日時を抜いたニュース本文（最初の2行）
        text = "".join(text[2:])

        if len(text) == min_len:
            print("="*50)
            print(len(text), category)
            print(text)
```

**[確認]**
- 最小テキストでもうまく単語を抽出できれば、うまく分類は可能

```python
# カテゴリ別テキスト数のヒストグラム
category_text_len = {}
for category, paths in all_category.items():
    
    one_category_text_len = []
    for i, path in enumerate(paths):
        with open(path) as f:
            text = f.readlines()
        text = "".join(text[2:])
        one_category_text_len.append(len(text))
    category_text_len[category] = one_category_text_len
```

```python
fig = plt.figure(figsize=(15,5))

ax = fig.add_subplot(1,2,1)
for category, text_len in category_text_len.items():
    ax.hist(text_len)
ax.set_xlabel("文字数")
ax.set_ylabel("frequency")
ax.set_title("全データ")

ax = fig.add_subplot(1,2,2)
for category, text_len in category_text_len.items():
    ax.hist([tl for tl in text_len if tl < 5000])
ax.set_xlabel("文字数")
ax.set_ylabel("frequency")
ax.set_title("5000文字以下")
```

**[確認]**
- `it-life-hack` `smax`の最大数は外れ値並み（5000文字以上はほぼ外れ値）
- カテゴリによってテキストの分布はかなり異なる

# 前処理

カテゴリ別のテキストデータフレームを作成する

```python
all_category_df = pd.DataFrame(columns=["category", "text"])
for category, paths in all_category.items():
     
    one_category_text = []
    for i, path in enumerate(paths):
        with open(path) as f:
            text = f.readlines()
        text = "".join(text[2:])
        one_category_text.append(text)
    
    one_category_df = pd.DataFrame(one_category_text, columns=["text"])
    one_category_df["category"] = category
    all_category_df = pd.concat([all_category_df, one_category_df])

all_category_df = all_category_df.reset_index(drop=True)
```

```python
all_category_df.head()
```

----
**処理**

1. unicode変換
1. 数字の削除
1. 空白の削除
1. URLの削除
1. 絵文字・顔文字の削除
1. 大文字の小文字統一
1. 文字の表記揺れを修正（名寄せ）
1. ストップワードの除去
1. [!][?]等の統一

```python
# unicode変換
all_category_df["text"] = all_category_df["text"].apply(
                lambda x: unicodedata.normalize("NFKC", x))
```

```python
replace_regex = {
    # 数字の削除
    "\d": "0",
    
    # 空白の削除
    "\u3000": "",
    
    # URLの削除
    "https?://[\w!\?/\+\+\-_~=;\.,\*&@#\$%\(\)'\[\]]+": ""
}
for regex, replace in replace_regex.items():
    pattern = re.compile(regex, re.MULTILINE+re.DOTALL)
    all_category_df["text"] = all_category_df["text"].apply(
                lambda x: pattern.sub(replace, x))
```
